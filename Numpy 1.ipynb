```python
import numpy as np
```

```python
# creating  1d numpy array
mylist1 = [101,102,103]
myarray = np.array(mylist1)
print(myarray)
```

```python
# creating a 2d numpy array
mylist2 = [101,102,103]
myarray2D = np.array([mylist1,mylist2])
print(myarray2D)
                
```

```python
# finding the dimension of the array - shape
print('myarray dimensions')
print(myarray.shape)
print('myarray 2D dimensions')
print(myarray2D.shape)

     
```

```python
#finding the datatype of the array -dtype
print('myarray datatype')
print(myarray.dtype)
print('myarray 2D datatype')
print(myarray2D.dtype)
```

```python
# array creation and intilization functions -zeros
zero_array = np.zeros (5)
print(zero_array)

zero_array2D = np.zeros([5,5])
print(zero_array2D)
                      
```

```python
# array creation and intilization functions -ones
print('1D')
zero_array = np.ones (5)
print(zero_array)
print('2D')
zero_array2D = np.ones([5,5])
print(zero_array2D)
                      
```

```python
# array creation and intilization functions -empty

print('1D')
zero_array = np.empty (5)
print(zero_array)
print('2D')
zero_array2D = np.empty([5,5])
print(zero_array2D)
            
```

```python
# array creation and intilization functions -eye

Identity_array = np.eye(3)
print(Identity_array)
    
```

```python
# array creation and intilization functions -arange 

Ap_array = np.arange(0,50,2)
print(Ap_array)
```
