```python
import numpy as np
```

```python
#scalar array multiplication- using * operator 

arr1 = np.array([[1,2,3],[5,6,7]])
print(arr1)
```

```python
arr2 = arr1*arr1
print(arr2)
```

```python
# Exponential multiplication using ** operator 
arr3 = arr1**2
print(arr3)
```

```python
# scalar subraction of array elements 

arr4 = arr3 - arr1
print(arr4)
```

```python
# scalar Divison of array elements 

arr5 = 1/arr1
print(arr5)
```
